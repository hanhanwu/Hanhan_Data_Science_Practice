```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from dtaidistance import dtw, ed
from scipy import stats
import similaritymeasures


%matplotlib inline  
```

```python
x1 = np.arange(0, 4*np.pi, 0.1)   # start,stop,step
x2 = np.arange(0, 5*np.pi, 0.1)  
x3 = np.arange(0, 4*np.pi, 0.5)

y1 = np.sin(x1)
y2 = -y1
y3 = y1+1
y4 = np.cos(x1)
y5 = np.sin(x2)
y6 = np.sin(x3)
y7 = np.sin(x1-0.5)
y8 = np.sin(x1-2)
```

```python
x1 = np.arange(0, 4*np.pi, 0.1)   # start,stop,step
x2 = np.arange(0, 5*np.pi, 0.1)  
x3 = np.arange(0, 4*np.pi, 0.5)

y1 = np.sin(x1)
y2 = -y1
y3 = y1+1
y4 = np.cos(x1)
y5 = np.sin(x2)
y6 = np.sin(x3)
y7 = np.sin(x1-0.5)
y8 = np.sin(x1-2)

y_lst = [y2, y3, y4, y5, y6, y7, y8]
yf_lst = ['-sin(x1)', 'sin(x1) + 1', 'cos(x1)', 'sin(x2)', 'sin(x3)', 'sin(x1-0.5)', 'sin(x-2)']

summary_dct = {}

for i in range(len(y_lst)):
    summary_dct[i] = {'s1': 'sin(x1)', 's2': yf_lst[i], 
                      'dtw': dtw.distance_fast(y1, y_lst[i]),
                      'pearson corr': max(stats.pearsonr(y1, y_lst[i])) if len(y1)==len(y_lst[i]) else None,
                      'frechet': similaritymeasures.frechet_dist(y1, y_lst[i])
                     }
    
summary_df = pd.DataFrame(summary_dct).T
summary_df = summary_df.sort_values(by='dtw')
summary_df
```

## Observations

* DTW (Dynamic Time Warping), Pearson Correlation and Frechet Distance can all be used to compare the similarity of 2 series
* Pearson Correlation require both series have the same length while DYW and Frechet don't require that
* To compare the shape similarity when 2 series seem to have larger vertial differences, using pearson correlation is better
* To compare the similarity when 2 series have larger horizontal differences, using dtw or frechet is better
* The way Frechet works makes it less sensitive to the changes between each 2 points of the series, and this might be why with Frechet, sin(x1) and sin(x3) have the smallest difference while in dtw, the smallest distance is in between sin(x1) and sin(x1-0.5)
